```python
import pandas as pd
import re
from io import StringIO
```

```python
with open('DA_raw.txt', 'r') as f:
    data = f.read()
data
```

```python
dates = []
times = []
senders = []
contents = []

current_date = None
current_time = None
current_sender = None
current_content = []

for line in data.splitlines():
    #match = re.match(r'(\d{2}/\d{2}/\d{2}), (\d{1,2}:\d{2} [ap]m) - (.*): (.*)', line)
    match = re.match(r"(\d{1,2}/\d{1,2}/\d{2}), (\d{1,2}:\d{2}) - (.*?): (.*?)(?=\n\d{1,2}/\d{1,2}/\d{2}, |\Z)", line)

    if match:
        if current_date:
            dates.append(current_date)
            times.append(current_time)
            senders.append(current_sender)
            contents.append(' '.join(current_content).strip())
        
        current_date, current_time, current_sender, line_content = match.groups()
        current_time = current_time.replace('\u202f',' ')
        current_content = [line_content.strip()]
        
    else:
        current_content.append(line.strip())

if current_date:
    dates.append(current_date)
    times.append(current_time)
    senders.append(current_sender)
    contents.append(' '.join(current_content).strip())

df = pd.DataFrame({
    'date': dates,
    'time': times,
    'sender': senders,
    'content': contents
})

df
```

```python
df.to_csv('DA_raw.csv', index=False)
```
